```python
# All the packages needed.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style('darkgrid')
%matplotlib inline
```

```python
#Checking if there is any null values in dataset
#The NULL value can cause problems when selecting data.
#However, because when comparing an unknown value to any other value, the result is always unknown and not included in the results

tweet_data.isna().sum()
```

```python
#Dropping the location column and user data as it is filled with null values.
#If columns have more than half of the rows as null then the entire column can be dropped.

tweet_data_drop = tweet_data.drop(['location','user_description'],axis = 1)
```

```python
tweet_data_drop.isna().sum()
```

```python
# PairPlot using seaborn

plt.figure(figsize=(11,6))
sns.pairplot(tweet_data_drop,diag_kind = 'kde', palette='husl', hue = 'searched_hashtag')
plt.show()
```

```python
# Plotting correlation matrix
correlation_map = tweet_data_drop.corr()
plt.figure(figsize=(30,15))
plt.title("Correlation Matrix of DataSet")
sns.heatmap(correlation_map,annot=True)
plt.show()
```

```python
#Dropping the verified column as it is not corelated to any other attributes.

tweet_data_drop = tweet_data_drop.drop(['verified'],axis = 1)
```

```python
# Plotting correlation matrix
correlation_map = tweet_data_drop.corr()
plt.figure(figsize=(30,15))
plt.title("Correlation Matrix of DataSet")
sns.heatmap(correlation_map,annot=True)
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(tweet_data_drop)
X_train_counts.shape
```

```python
count_vect.vocabulary_.get(u'searched_hashtag')
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
X_train_tf.shape
```
