```python
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.feature_selection import SelectKBest, chi2

X = wine_data.drop(columns=['quality']) #wine_data.drop(columns=['quality', 'sulphates', 'citric acid', 'free sulfur dioxide'])
y = wine_data['quality']
ros = SMOTE(random_state=42)
X_res, y_res = ros.fit_resample(X, y)
print('Resampled dataset shape %s' % Counter(y_res))

X_res = SelectKBest(chi2, k=8).fit_transform(X_res, y_res)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, plot_confusion_matrix, plot_roc_curve

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.1, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, shuffle=False, random_state=42)
print("Training X:" , X_train.shape)
print("Training y:" , y_train.shape)
print("Test X:" , X_test.shape)
print("Test y:" , y_test.shape)
print("Validation X:" , X_val.shape)
print("Validation y:" , y_val.shape)

svc = SVC()
svc.fit(X_train, y_train)
predictions = svc.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print('Accuracy:', accuracy)
print(classification_report(y_test, predictions))

plot_confusion_matrix(svc, X_train, y_train, normalize='pred')
plt.title('Train set')
plt.show()
plot_confusion_matrix(svc, X_test, y_test, normalize='pred')
plt.title('Test set')
plt.show()
plot_roc_curve(svc, X_test, y_test)
plt.show()
```

```python
wine_data = pd.read_csv('/content/drive/MyDrive/SFML/winequality-white.csv', delimiter=';')
wine_data.head()
```

```python
print(wine_data['quality'].value_counts())
quality = wine_data['quality']
wine_data = wine_data.drop(columns='quality').apply(lambda iterator: ((iterator.max() - iterator)/(iterator.max() - iterator.min())).round(2))
wine_data['quality'] = quality
wine_data.head()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(wine_data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import classification_report, mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(wine_data.drop(columns=['quality', 'sulphates', 'citric acid', 'free sulfur dioxide']), wine_data['quality'], test_size=20, shuffle=False, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=25, shuffle=False, random_state=42)


rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)


print("R-squared:", rfr.score(X_test, y_test))

predictions = rfr.predict(X_test)
print(predictions)

ypred = rfr.predict(X_test)

mse = mean_squared_error(y_test, ypred)
rmse = mean_squared_error(y_test, ypred, squared=False)
print("MSE: ", mse)
print("RMSE: ", rmse)
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, linewidth=1, label="original")
plt.plot(x_ax, ypred, linewidth=1.1, label="predicted")
plt.title("y-test and y-predicted data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
# Random search of parameters, using 3 fold cross validation,
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rfr, param_distributions = random_grid, n_iter = 10, cv = 10, verbose=3, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_val, y_val)

print(rf_random.best_params_)
print(rf_random.best_estimator_)
```

```python
rfr = RandomForestRegressor(max_depth=30, max_features='sqrt', min_samples_split=5, n_estimators=1400)
rfr.fit(X_train, y_train)


print("R-squared:", rfr.score(X_test, y_test))

predictions = rfr.predict(X_test)
print(predictions)

ypred = rfr.predict(X_test)

mse = mean_squared_error(y_test, ypred)
rmse = mean_squared_error(y_test, ypred, squared=False)
print("MSE: ", mse)
print("RMSE: ", rmse)
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, linewidth=1, label="original")
plt.plot(x_ax, ypred, linewidth=1.1, label="predicted")
plt.title("y-test and y-predicted data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```
